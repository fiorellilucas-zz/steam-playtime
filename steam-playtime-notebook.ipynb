```python
import pandas as pd
import numpy as np
```

```python
df_steam = pd.read_csv("steam-playtime-2021-06-07.csv", index_col="game_title")
df_steam
```

```python
df_hltb = pd.read_csv("hltb-playtime-2021-06-07.csv", index_col="game_title")
df_hltb
```

```python
df_games = pd.concat([df_steam, df_hltb], axis=1).dropna()
df_games
```

```python
df_games["time_to_beat"].replace("-", np.nan, inplace=True)
```

```python
df_games.dropna(inplace=True)
```

```python
df_games["time_to_beat"] = df_games["time_to_beat"].astype("float64")
```

```python
df_games.info()
```

```python
df_games = df_games.assign(played_ratio=(df_games["playtime_forever"]/df_games["time_to_beat"]) * 100)
```

```python
df_games.sort_values("played_ratio", ascending=False).head(40)
```

```python
print(f'Average played = {round(df_games["played_ratio"].mean(), 2)}%')
```
